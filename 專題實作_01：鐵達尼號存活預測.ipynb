```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv'
df = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```

```python
df
#乘客編號、是否存活（0=沒有、1=有）Pclass船艙等級(1=Upper,2=Middle,3=Lower)
#名字、性別、年齡（小於1歲使用分數表示；年齡是預測的話會以XX.5表示）、sibsp是否有手足及配偶、parch是否有直系血親（若是與保母旅遊的小孩，parch為0）


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the relationship between survived and sex
sns.barplot(x="Sex", y="Survived", data=df)
plt.show()

# Plot the relationship between survived and age
sns.barplot(x="Age", y="Survived", data=df)
plt.show()

# Plot the relationship between survived and pclass
sns.barplot(x="Pclass", y="Survived", data=df)
plt.show()


##存活率高相關連的特徵：女性、壯年、艙別愈高存活率愈高
```
