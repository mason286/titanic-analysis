```python
import pandas as pd
import numpy as np

d = [
    {'city': 'Austin', 'visitor': 139, 'weekday': 'Sun'},
    {'city': 'Dallas', 'visitor': 237, 'weekday': 'Sun'},
    {'city': 'Austin', 'visitor': 326, 'weekday': 'Mon'},
    {'city': 'Dallas', 'visitor': 456, 'weekday': 'Mon'}
]
```

```python
df=pd.DataFrame(d)
grouped = df.groupby('weekday')
result = grouped.agg({'visitor': 'sum'})
print(result)
```

```python
import datetime
import pandas as pd
from bs4 import BeautifulSoup
import requests
import sqlite3
conn = sqlite3.connect('mydatabase.db')

content = requests.get("https://rate.bot.com.tw/xrt?Lang=zh-TW")
soup = BeautifulSoup(content.text,"html.parser")   #指定使用解析html的參數   html5lib html.parser

currency = soup.find_all('div', class_='visible-phone print_hide')

cash_rate_buy = soup.find_all('td', class_='rate-content-cash text-right print_hide', attrs={'data-table': '本行現金買入'})
cash_rate_sell = soup.find_all('td', class_='rate-content-cash text-right print_hide', attrs={'data-table': '本行現金賣出'})
spot_rate_buy = soup.find_all('td', class_='rate-content-sight text-right print_hide', attrs={'data-table': '本行即期買入'})
spot_rate_sell = soup.find_all('td', class_='rate-content-sight text-right print_hide', attrs={'data-table': '本行即期賣出'})

data = []
# .strip() 移除文本內容中的空白字符（空格、換行符等）
for i in range(len(currency)):
    currency_name = currency[i].text.strip()
    cash_buy = cash_rate_buy[i].text.strip()
    cash_sell = cash_rate_sell[i].text.strip()
    spot_buy = spot_rate_buy[i].text.strip()
    spot_sell = spot_rate_sell[i].text.strip()
    data.append([currency_name, cash_buy, cash_sell, spot_buy, spot_sell])


data
```

```python

df = pd.DataFrame(data, columns=['幣別', '現金匯率本行買入', '現金匯率本行賣出', '即期匯率本行買入', '即期匯率本行賣出'])

df["更新日期"] = datetime.datetime.now()

print(df)
```

```python
df.to_sql('exchange_rates',conn,if_exists='append',index=False)
```

```python
c=conn.cursor()
cursor=c.execute("select * from exchange_rates")
```

```python
for row in cursor:
    print("幣別 =", row[0])
    print("現金匯率本行買入 =", row[1])
    print("現金匯率本行賣出 =", row[2])
    print("即期匯率本行買入 =", row[3])
    print("即期匯率本行賣出 =", row[4])
    print("更新日期 =", row[5], "\n")
```
